```python
print(json.dumps(te_result, indent=4))  # Pretty-print the output JSON
```

```python

# Initialize node and edge lists
nodes = []
edges = []

# Helper function to check if a node already exists
def add_node(node_id, node_type, importance=0.5):
    if not any(node["id"] == node_id for node in nodes):
        nodes.append({"id": node_id, "type": node_type, "importance": importance})

# Extract nodes from themes and subthemes
for theme in te_result["themes"]:
    add_node(theme["name"], "theme", theme["importance"])
    
    # Extract subthemes as nodes and create edges connecting them to their parent themes
    for subtheme in theme.get("subthemes", []):
        add_node(subtheme["name"], "subtheme", subtheme["importance"])
        edges.append({
            "source": theme["name"],
            "target": subtheme["name"],
            "name": "subtheme of",
            "type": "thematic",
            "importnace": 1  # Default weight for theme-subtheme relationships
        })

# Extract edges from entity relationships
for entity in er_result["entities"]:
    add_node(entity["name"], entity["type"], entity["importance"])

# Extract edges from relationships
for relation in er_result["relationships"]:
    edges.append({
        "source": relation["subject"],
        "target": relation["object"],
        "name": relation["relation"],
        "type": "correlated",
        "importance": 1  # Default weight, you can modify this based on confidence if available
    })

# Extract edges from causal inference
for relation in ci_result["causal_relations"]:
    edges.append({
        "source": relation["cause"],
        "target": relation["effect"],
        "name": "",
        "type": "causal",
        "importnance": relation["confidence"]
    })


# Combine everything into final JSON format
graph_json = {
    "nodes": nodes,
    "edges": edges
}

# Save to a JSON file
with open("narrative_genome.json", "w") as f:
    json.dump(graph_json, f, indent=4)

# Print output for verification
print(json.dumps(graph_json, indent=4))
```

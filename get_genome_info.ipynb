```python

# Initialize node and edge lists
nodes = []
edges = []

# Helper function to check if a node already exists
def add_node(node_id, node_type, importance=0.5):
    if not any(node["id"] == node_id for node in nodes):
        nodes.append({"id": node_id, "type": node_type, "importance": importance})

# Extract nodes from themes and subthemes
for theme in te_result["themes"]:
    add_node(theme["name"], "theme", theme["importance"])
    
    # Extract subthemes as nodes and create edges connecting them to their parent themes
    for subtheme in theme.get("subthemes", []):
        add_node(subtheme["name"], "subtheme", subtheme["importance"])
        edges.append({
            "source": theme["name"],
            "target": subtheme["name"],
            "name": "parent theme of",
            "type": "thematic",
            "importance": subtheme["importance"]  # Default weight for theme-subtheme relationships
        })

# Extract edges from entity relationships
for entity in er_result["entities"]:
    add_node(entity["name"], entity["type"], entity["importance"])

# Extract edges from relationships
for relation in er_result["relationships"]:
    edges.append({
        "source": relation["subject"],
        "target": relation["object"],
        "name": relation["relation"],
        "type": "correlated",
        "importance": 5  # Default weight, you can modify this based on confidence if available
    })


print(nodes)

```

```python
from sentence_transformers import SentenceTransformer, util

def semantic_match(query, node_data, threshold=0.4):
    """Matches query to nodes based on semantic similarity."""
    model = SentenceTransformer('all-MiniLM-L6-v2')
    query_embedding = model.encode(query, convert_to_tensor=True)
    matches = []
    for node in node_data: #access nodes from the dictionary.
        node_id = node["id"]
        node_embedding = model.encode(node_id, convert_to_tensor=True)
        similarity = util.cos_sim(query_embedding, node_embedding).item()
        if similarity > threshold:
            matches.append((node_id, similarity))
    sorted_matches = sorted(matches, key=lambda item: item[1], reverse=True)
    return sorted_matches

def process_cause_effect(cause_effect_data, node_data):
    """Processes cause and effect data to find matching nodes using only semantic matching."""
    cause_node = None
    effect_node = None

    cause_matches = semantic_match(cause_effect_data["cause"], node_data)
    effect_matches = semantic_match(cause_effect_data["effect"], node_data)

    if cause_matches:
        cause_node = cause_matches[0][0]  # Get the top match
    if effect_matches:
        effect_node = effect_matches[0][0]  # Get the top match

    return cause_node, effect_node
```

```python
# Extract edges from causal inference
for relation in ci_result["causal_relations"]:
    
    cause_node, effect_node = process_cause_effect(relation, nodes)
    # print(f"Cause Node: {cause_node}, Effect Node: {effect_node}\n")
    if cause_node is not None and effect_node is not None:
        edges.append({
            "source": cause_node,
            "target": effect_node,
            "name": relation["proof"],
            "type": "causal",
            "importance": relation["confidence"]
        })

print(edges)
```

```python
# Combine everything into final JSON format
graph_json = {
    "nodes": nodes,
    "edges": edges
}

# Save to a JSON file
with open("narrative_genome_amazon.json", "w") as f:
    json.dump(graph_json, f, indent=4)

# Print output for verification
print(json.dumps(graph_json, indent=4))
```
